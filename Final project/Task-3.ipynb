### Задание 3

___


В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

___

#### <span style="text-decoration: underline">_Обычная механика:_ </span>


* **Installs** - посчитать количество установок приложения в тематические дни и в обычные, так как обычно такие события так же привлекают новых пользователей

* **Retention rate** - сравнить ретеншен 7го дня в тематические и в обычные дни

* **ASD** - среднее время сессий за период времени.

* **DAU/WAU** - посмотреть на количество активных пользователей в тематические дни/недели и в обычные.

* **Stickiness rate**  - посмотреть отношение DAU / MAU в тематические дни/недели и в обычные.
Относительная и абсолютная конверсии - позволят определить как проведение тематических событий сказывавается на привлечении новой аудитории.

* **ARPU/ARPPU** - увеличиваются ли показатели во время проведения тематических событий


#### <span style="text-decoration: underline">_Измененная механика:_</span>

Судя по описанию измененная механика несет негативный пользовательский опыт, так как многие не любят проигрывать и терять достижения, следовательно и метрики будут измерять значения связаные с этим:

* **Churn rate после неудачной попытки** - позволит оценить влияние данной механики на то будет ли человек продолжать возвращаться в игру

* **ASD после неудачной попытки** - позволит оценить влияние данной механики на то, будет ли человек продолжать играть

* **Среднее кол-во неудачных попыток на пользователя** - позволит понять на сколько в среднем у пользователей хватает терпения играть в такой режим
